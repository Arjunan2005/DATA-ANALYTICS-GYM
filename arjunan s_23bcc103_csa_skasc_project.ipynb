```python
import pandas as pd

# Load the dataset
file_name = "gym_member_exercise_tracking.csv"
df = pd.read_csv (file_name)

# Preview the data
print(df.head())
print(df.info())
print(df.describe())
```

```python
#Check for missing values
print("Missing values in each column:")
print(df.isnull().sum())

# Get the actual column names from the DataFrame
actual_columns = df.columns

# Fill missing values or drop rows/columns
# Check if 'Rating' and 'Company' are in actual columns before filling NaNs
if 'Rating' in actual_columns:
    df['Rating'] = df['Rating'].fillna(df['Rating'].mean())  # Example for numeric column
if 'Company' in actual_columns:
    df['Company'] = df['Company'].fillna('Unknown')  # Example for categorical column
else:
    # If 'Company' is not found, print an error or handle it appropriately
    print("Column 'Company' not found in the DataFrame.")
    # You might want to check for similar column names or adjust your code accordingly
```

```python
# Check for duplicates
print("Number of duplicate rows:", df.duplicated().sum())

# Drop duplicate rows
df = df.drop_duplicates()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# Load the cleaned dataset
 # Update the file path with the correct location

# Try to read the dataset and handle file not found error
try:
    df = pd.read_csv('/content/gym_member_exercise_tracking.csv')
except FileNotFoundError:
    print("Dataset not found. Please check the file path.")
    exit()

# Check for required columns in the dataset
required_columns = ['ExerciseType', 'Duration(min)', 'CaloriesBurned', 'IntensityLevel', 'Date', 'MemberID']
missing_columns = [col for col in required_columns if col not in df.columns]
if missing_columns:
    print(f"Missing columns: {', '.join(missing_columns)}. Please ensure the dataset has all required columns.")
    exit()

# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df.dropna(subset=['Date'], inplace=True)

# 1. Exercise Popularity
exercise_popularity = df['ExerciseType'].value_counts(normalize=True) * 100
print("\nExercise Popularity (%):")
print(exercise_popularity)

# Plot Exercise Popularity
plt.figure(figsize=(10, 6))
exercise_popularity.plot(kind='bar', color='skyblue')
plt.title('Exercise Popularity (%)')
plt.xlabel('Exercise Type')
plt.ylabel('Percentage')
plt.xticks(rotation=45)
plt.show()
```

```python
# 2. Average Duration and Calories Burned by Exercise Type
avg_metrics = df.groupby('ExerciseType')[['Duration(min)', 'CaloriesBurned']].mean()
print("\nAverage Duration and Calories Burned per Exercise Type:")
print(avg_metrics)

# Plot Average Duration and Calories Burned
avg_metrics.plot(kind='bar', figsize=(12, 6))
plt.title('Average Duration and Calories Burned by Exercise Type')
plt.ylabel('Average Value')
plt.xticks(rotation=45)
plt.legend(title='Metrics')
plt.show()
```

```python
# 3. Intensity Preferences
intensity_distribution = df['IntensityLevel'].value_counts(normalize=True) * 100
print("\nIntensity Preferences (%):")
print(intensity_distribution)

# Plot Intensity Preferences
plt.figure(figsize=(8, 6))
intensity_distribution.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['lightgreen', 'gold', 'lightcoral'])
plt.title('Intensity Preferences (%)')
plt.ylabel(None)
plt.show()
```

```python
# 4. Daily Activity Trends
daily_sessions = df.groupby('Date').size()
print("\nDaily Activity Trends:")
print(daily_sessions.head())

# Plot Daily Activity Trends
plt.figure(figsize=(14, 6))
plt.plot(daily_sessions.index, daily_sessions.values, marker='o', linestyle='-', color='b')
plt.title('Daily Activity Trends')
plt.xlabel('Date')
plt.ylabel('Number of Sessions')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

```

```python
# 5. Member Behavior Insights
# Most Active Members
member_activity = df.groupby('MemberID').size().sort_values(ascending=False)
print("\nMost Active Members:")
print(member_activity.head())

# Calorie Efficiency Calculation and Display
df['CaloriesPerMinute'] = df['CaloriesBurned'] / df['Duration(min)']
print("\nCalorie Efficiency (Top 5):")
print(df[['MemberID', 'ExerciseType', 'CaloriesPerMinute']]
      .sort_values(by='CaloriesPerMinute', ascending=False)
      .head())

```

```python
# 6. Equipment Usage Trends
equipment_exercises = ['Treadmill', 'Cycling', 'Weightlifting']
equipment_usage = df[df['ExerciseType'].isin(equipment_exercises)].groupby('Date').size()
print("\nEquipment Usage Trends:")
print(equipment_usage.head())

# Plot Equipment Usage Trends
plt.figure(figsize=(14, 6))
plt.plot(equipment_usage.index, equipment_usage.values, marker='o', linestyle='-', color='orange')
plt.title('Equipment Usage Trends')
plt.xlabel('Date')
plt.ylabel('Number of Sessions')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python

  # 7. Histogram: Distribution of duration
  import seaborn as sns
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x="Duration(min)", bins=20, kde=True)
plt.title("Distribution of Exercise Duration")
plt.tight_layout()
plt.show()
```

```python
# 8. Scatter Plot: Relationship between duration and calories burned
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="Duration(min)", y="CaloriesBurned", hue="IntensityLevel", alpha=0.7)
plt.title("Duration vs. Calories Burned by Intensity Level")
plt.tight_layout()
plt.show()
```

```python
# 9. Box Plot: Duration by intensity level
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="IntensityLevel", y="Duration(min)")
plt.title("Duration by Intensity Level")
plt.tight_layout()
plt.show()
```

```python
# 10. Heatmap: Correlation between numerical features
plt.figure(figsize=(10, 8))
sns.heatmap(df[["Duration(min)", "CaloriesBurned"]].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Between Numerical Features")
plt.tight_layout()
plt.show()
```

```python
# 11. Violin Plot: Distribution of calories burned by exercise type
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x="ExerciseType", y="CaloriesBurned", split=True)
plt.title("Calories Burned by Exercise Type")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**INSIGHTS**
1. **Most Popular Exercise Types**:  
   Identify the exercise type most commonly performed by gym members. This can help the gym allocate resources effectively (e.g., more treadmills if running is most popular).

2. **Calories Burned Efficiency**:  
   Analyze which exercise types have the highest calories burned per minute on average. This insight can guide members who aim for quick, high-impact workouts and inform fitness program design.

3. **Intensity and Duration Relationship**:  
   Evaluate the relationship between intensity levels (Low, Medium, High) and workout duration. For example, high-intensity workouts might have shorter durations but higher calories burned per session, offering valuable insights for fitness goal planning.



**CONCLUSION**

Analyzing the gym members' exercise data provides actionable insights to enhance both gym operations and member experiences. The most popular exercises can be identified, enabling the gym to allocate resources and prioritize equipment based on member preferences. Additionally, understanding which exercise types burn the most calories per minute allows for the design of effective fitness plans, particularly for members focusing on weight loss or maximizing calorie burn. Furthermore, examining the relationship between intensity levels and workout duration helps tailor recommendations for members seeking to balance workout efficiency and endurance. These insights collectively support the optimization of workout programs, improved member satisfaction, and efficient resource utilization.
